```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class Queue:
    def __init__(self):
        self.root = []

    def enqueue(self, value):
        self.root.append(value)
        return True

    def dequeue(self):
        if not self.root:
            return None
        return self.root.pop(0)


class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        new_node = Node(value)
        if self.root is None:
            self.root = new_node
            return True
        temp = self.root
        while True:
            if new_node.value == temp.value:
                return False
            if new_node.value < temp.value:
                if temp.left is None:
                    temp.left = new_node
                    return True
                temp = temp.left
            else:
                if temp.right is None:
                    temp.right = new_node
                    return True
                temp = temp.right

    def contains(self, value):
        if self.root is None:
            return False
        temp = self.root
        while (temp):
            if value < temp.value:
                temp = temp.left
            if value > temp.value:
                temp = temp.right
            else:
                return True
        return False
    
    def original_BFS(self):
        current_node = self.root
        queue = Queue()
        queue.enqueue(self.root)
        while queue.root:
            current_node = queue.dequeue()
            if current_node is not None:
                print(current_node.value)
                if current_node.left is not None:
                    queue.enqueue(current_node.left)
                if current_node.right is not None:
                    queue.enqueue(current_node.right)
            else:
                break
        
    def dfs_pre_order(self):
        result = []
        def traverse(current_node):
            result.append(current_node.value)
            if current_node.left is not None:
                traverse(current_node.left)
            if current_node.right is not None:
                traverse(current_node.right)
        traverse(self.root)
        return result
    
    def dfs_post_order(self):
        result = []
        def traverse(current_node):
            if current_node.left is not None:
                traverse(current_node.left)
            if current_node.right is not None:
                traverse(current_node.right)
            result.append(current_node.value)
        traverse(self.root)
        return result
    
    def dfs_in_order(self):
        result = []
        def traverse(current_node):
            if current_node.left is not None:
                traverse(current_node.left)
            result.append(current_node.value)
            if current_node.right is not None:
                traverse(current_node.right)
        traverse(self.root)
        return result
    
    def find_my_parent(self, value):
        if self.root is None or self.root.value == value:
            return False, None
        parent = None
        current_node = self.root
        while current_node:
            if value < current_node.value:
                parent, current_node = current_node, current_node.left
            elif value > current_node.value:
                parent, current_node = current_node, current_node.right
            else:
                return True, parent.value
        return False, None

my_tree = BinarySearchTree()
my_tree.insert(47)
my_tree.insert(21)
my_tree.insert(76)
my_tree.insert(18)
my_tree.insert(27)
my_tree.insert(52)
my_tree.insert(82)

my_tree.original_BFS()
```
