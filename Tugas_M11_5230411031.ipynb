```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.len = 0

    def insert(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node
        self.len += 1

    def display(self):
        current = self.head
        while current != None:
            print(current.value, end=' ')
            current = current.next
        print()

    def asc_linked_list(self):
        for i in range(self.len - 1):
            current = self.head
            for j in range(self.len - 1):
                if current.value > current.next.value:
                    temp = current.value
                    current.value = current.next.value
                    current.next.value = temp
                current = current.next
    
    def descend_linked_list(self):
        for i in range(self.len - 1):
            current = self.head
            for j in range(self.len - 1):
                if current.value < current.next.value:
                    temp = current.value
                    current.value = current.next.value
                    current.next.value = temp
                current = current.next
    
# main = LinkedList()
# main.insert(5)
# main.insert(3)
# main.insert(8)
# main.insert(1)
# main.insert(2)
# print("Original linked list:")
# main.display()
# main.asc_linked_list()
# print("Ascending linked list:")
# main.display()
# main.descend_linked_list()
# print("Descending linked list:")
# main.display()

    def menu():
        print("1. Insert")
        print("2. Display linked list")
        print("3. Ascending linked list")
        print("4. Descending linked list")
        print("5. Exit")
        

def main():
    main = LinkedList()
    LinkedList.menu()
    while True:
        choice = int(input("Enter your choice: "))
        if choice == 1:
            while True:
                value = int(input("Enter value to insert (insert -1 to finish): "))
                if value == -1:
                    break
                else :
                    main.insert(value)
                    print("Value", value, "inserted.")
        elif choice == 2:
            print("Original linked list:")
            main.display()
        elif choice == 3:
            print("Ascending linked list:")
            main.asc_linked_list()
            main.display()
        elif choice == 4:
            print("Descending linked list:")
            main.descend_linked_list()
            main.display()
        elif choice == 5:
            break

main()
```
