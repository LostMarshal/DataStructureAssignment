```python
class Graph:
    def __init__(self):
        self.adj_list = {}
    
    def print_graph(self):
        for vertex in self.adj_list:
            print(vertex, ":", self.adj_list[vertex])

    def add_vertex(self, vertex):
        if vertex not in self.adj_list.keys():
            self.adj_list[vertex] = []
            return True
        return False
    
    def add_edge(self, v1, v2):
        if v1 in self.adj_list.keys() and v2 in self.adj_list.keys():
            self.adj_list[v1].append(v2)
            self.adj_list[v2].append(v1)
            return True
        return False
    
    def remove_edge(self, v1, v2):
        if v1 in self.adj_list.keys() and v2 in self.adj_list.keys():
            try:
                self.adj_list[v1].remove(v2)
                self.adj_list[v2].remove(v1)
            except ValueError:
                pass
            return True
        return False
    
    def remove_vertex(self, vertex):
        if vertex in self.adj_list.keys():
            for oter_vertex in self.adj_list[vertex]:
                self.adj_list[oter_vertex].remove(vertex)
            del self.adj_list[vertex]
            return True
        return False
    
    def isInGraph(self, vertex):
        if vertex in self.adj_list.keys():
            return True
        return False
    
    def printAllConnected(self, vertex):
        if vertex in self.adj_list.keys():
            print(vertex, ":", self.adj_list[vertex])
        else:
            print('Vertex not in graph')

my_graph = Graph()
# my_graph.add_vertex(82)
# my_graph.add_vertex(27)
# my_graph.add_vertex(52)
# my_graph.add_vertex(18)
# my_graph.add_vertex(21)
# my_graph.add_vertex(76)
# my_graph.add_vertex(44)

# my_graph.add_edge(82, 27)
# my_graph.add_edge(27, 52)
# my_graph.add_edge(52, 18)
# my_graph.add_edge(18, 21)
# my_graph.add_edge(21, 76)
# my_graph.add_edge(76, 44)
# my_graph.add_edge(44, 18)
# my_graph.add_edge(44, 21)

# print('Graph :')
# my_graph.print_graph()

# print('\nCheck if 82 is in the graph :', my_graph.isInGraph(82))
# print('Check if 99 is in the graph :', my_graph.isInGraph(99))

# print('\nPrint all connected to 44 :')
# my_graph.printAllConnected(44)

# print('\nRemove all connected to 35 :')
# my_graph.printAllConnected(35)

while True:
    print("Menu :")
    print("1. Add vertex")
    print("2. Add edge")
    print("3. Remove vertex")
    print("4. Remove edge")
    print("5. Print graph")
    print("6. Print all connected")
    print("7. Check if vertex is in graph")
    print("8. Exit")
    choice = int(input("Enter your choice : "))
    if choice == 1:
        while True:
            print ("Enter -1 to exit")
            vertex = int(input("Enter vertex : "))
            if vertex == -1:
                break
            if my_graph.add_vertex(vertex):
                print("Vertex added")
            else:
                print("Vertex already exists")
    elif choice == 2:
        while True:
            print ("Enter -1 to exit")
            v1 = int(input("Enter vertex 1 : "))
            v2 = int(input("Enter vertex 2 : "))
            if my_graph.add_edge(v1, v2):
                print("Edge added")
            elif v1 == -1 or v2 == -1:
                break
            else:
                print("Invalid vertices")
    elif choice == 3:
        vertex = int(input("Enter vertex : "))
        if my_graph.remove_vertex(vertex):
            print("Removed vertex : ", vertex)
        else:
            print("Vertex not in graph")
    elif choice == 4:
        v1 = int(input("Enter vertex 1 : "))
        v2 = int(input("Enter vertex 2 : "))
        if my_graph.remove_edge(v1, v2):
            print("Removed edge: ", v1, "-", v2)
        else:
            print("Invalid vertex")
    elif choice == 5:
        my_graph.print_graph()
    elif choice == 6:
        vertex = int(input("Enter vertex : "))
        my_graph.printAllConnected(vertex)
    elif choice == 7:
        vertex = int(input("Enter vertex : "))
        if my_graph.isInGraph(vertex):
            print("Vertex is in graph")
        else:
            print("Vertex is not in graph")
    elif choice == 8:
        break
    else:
        print("Invalid choice")
    print()
```
